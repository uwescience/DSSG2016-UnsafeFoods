```python
from urllib.request import urlopen
from urllib.error import HTTPError
from bs4 import BeautifulSoup
import re
import pandas
```

```python
recall_data = pandas.read_csv("C:/Users/cvint/Desktop/DSSG Unsafe Foods/FDA_recalls.csv",skip_blank_lines = True,encoding='ISO-8859-1')
```

```python
upc_col = []

for rownum in range(recall_data.shape[0]):
    print(rownum)
    link = recall_data.COMPANY_RELEASE_LINK[rownum]
    #print(str(rownum))
    upc_col.append([])
    try:
        upc_url = urlopen(link)
        upc_bs = BeautifulSoup(upc_url,"html.parser")
    
        #get all upc codes
        #print(upc_bs.get_text())
        upcs = re.findall('\D(\d{12})\D|\D(\d[\s-]\d{5}[-\s]?\d{5}[-\s]\d)\D|\D(\d{6}[-\s]\d{5}[-\s]?\d)\D|\D(\d{5}[\s-]?\d{5})\D|\D(\d{6}[\s-]\d{5})\D',str(upc_bs))
        #print(upcs)
        for upc in upcs:
            #print(upc)
            for s in upc:
                #print(s)
                if s != '':
                    upc_col[rownum].append(s)
            
                
    except HTTPError:
        upc_col[rownum].append('url not found')
```

```python
for upc_set in upc_col:
    for index in range(len(upc_set)):
        upc_set[index] = upc_set[index].replace('-','')
        upc_set[index] = upc_set[index].replace(' ','')
    upc_set = list(set(upc_set))
```

```python
recall_data['upcs'] = pandas.Series(upc_col, index=recall_data.index)
```

```python
recall_data_to_save = recall_data[['DATE','BRAND_NAME','PRODUCT_DESCRIPTION','REASON','COMPANY','COMPANY_RELEASE_LINK','PHOTOS_LINK']]
recall_data_to_save['upcs'] = pandas.Series([None] * recall_data_to_save.shape[0], index=recall_data_to_save.index)
recall_data_to_save['asins'] = pandas.Series([None] * recall_data_to_save.shape[0], index=recall_data_to_save.index)
for row in range(recall_data.shape[0]):
    upc_str = ''
    for upc in recall_data.upcs[row]:
        upc = upc.replace(' ','')
        upc = upc.replace('-','')
        upc_str = upc_str + ';' + str(upc).strip()
    recall_data_to_save.upcs[row] = upc_str[1:]
    asin_str = ''
    for asin in recall_data.asins[row]:
        if asin not in ['UPCNOTFOUND','UPClength-0']:
            asin_str = asin_str + ';' + asin
    recall_data_to_save.asins[row] = asin_str[1:]
```
