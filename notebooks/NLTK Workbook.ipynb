```python
import pandas
import nltk
import re
from nltk.corpus import PlaintextCorpusReader
```

```python
review_data = pandas.read_csv('review_data.csv')
```

```python
import os

test_corpus = review_data.reviewText

# Make new dir for the corpus.
corpusdir = 'mikescorpus/'
if not os.path.isdir(corpusdir):
    os.mkdir(corpusdir)

# Output the files into the directory.
filename = 0
for text in test_corpus:
    filename+=1
    file = open(corpusdir + str(filename)+ '.txt','w')
    file.write(text)
    file.close()
```

```python
corpus_root = 'reviews/'
```

```python
review_corpus = PlaintextCorpusReader(corpus_root, ".*")
```

```python
review_corpus.words()
```

```python
import nltk.data
from nltk.tokenize import sent_tokenize, word_tokenize

tokenizer = nltk.data.load('tokenizers/punkt/english.pickle')
token_text = []
for infile in sorted(review_corpus.fileids())[:1000]:
    fin = review_corpus.open(infile)
    token_text.append(word_tokenize(fin.read().strip()))
```

```python
from nltk.stem.lancaster import LancasterStemmer
lancaster = LancasterStemmer()
```

```python
for tokens in token_text:
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
        tokens[i] = lancaster.stem(tokens[i])
        tokens[i] = re.sub('[\d\._]','',tokens[i])
```

```python
token_text[0]
```

```python
fixed_docs = []
for i in token_text:
    string = i[0]
    for s in range(1,len(i)):
        string = string + " " + i[s]
    fixed_docs.append(string)  
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction import text   

vectorizer = CountVectorizer(fixed_docs,stop_words = text.ENGLISH_STOP_WORDS)
x1 = vectorizer.fit_transform(fixed_docs)

#create dataFrame
df = pd.DataFrame(x1.toarray().transpose(), index = vectorizer.get_feature_names())
```
